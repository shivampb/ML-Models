```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/restaurant_data.csv')
df.head(1)
```

**Data Analysis**

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
# Group by City and Average Cost
plt.figure(figsize=(12,6))
df.groupby("City")["Average Cost for two"].mean().sort_values(ascending=False).head().plot(kind='area',color='orange')
plt.title("Average Cost For Two Per City, Top 5")
plt.xlabel('City')
plt.ylabel('Average res Price')
```

```python
plt.figure(figsize=(12,6))
df.groupby('Cuisines')["Votes"].sum().sort_values(ascending=False).head(10).plot(kind='line',color="pink")
plt.title("Votes Based on Cuisines")
plt.xlabel("Cuisines")
plt.ylabel("Total Votes")
```
